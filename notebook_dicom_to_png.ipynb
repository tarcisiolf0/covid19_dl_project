Função para armazenar os caminhos das imagens da base de dados do Kagle

```python
def get_dicom_paths(path):
    dicom_paths = []
    for root, dirnames, filenames in os.walk(path):
        for filename in filenames:
            if filename.endswith(".dcm"):
                dicom_paths.append(os.path.join(root, filename))
    return dicom_paths
```

Função para converter as imagens de dycom para png

```python
def diacom_to_png(file_path):
    # Read a DICOM dataset stored in accordance with the DICOM File Format.
    dimg = pydicom.dcmread(file_path)
    
    # Copy of the array, cast to a specified type.
    im = dimg.pixel_array.astype(float)
    
    rescaled_image = (np.maximum(im,0)/im.max())*255
    
    final_image = np.uint8(rescaled_image)
    
    # Creates an image memory from an object exporting the array interface (using the buffer protocol).
    final_image = Image.fromarray(final_image)
    
    return final_image
```

```python
root = '/home/tarcisio/Downloads/dataset_dcm/train'
diacom_paths = get_dicom_paths(root)
print(diacom_paths)
```

```python
for file_path in diacom_paths:
    png_img = diacom_to_png(file_path)
    
    save_path = '/home/tarcisio/Downloads/dataset_dcm/images_png/'
    file_name, ext = os.path.splitext(os.path.basename(file_path))
    
    png_img.save(save_path+file_name+'.png')
```
